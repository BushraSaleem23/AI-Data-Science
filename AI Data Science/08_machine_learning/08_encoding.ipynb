## Feature Encoding

Feature encoding is the process of transforming categoricl fatures to numerical features.This is neessay because ML algoriths can only handle numeric features.There aare many different ways to encoe cat features and each method has its own advantges and disadvantages.I this notebook ,we will exlore some of the most popular methods for encoding cat features.

### Types of Feature Encoding

- One-Hot Encoding
- Label encoding
- ordinal encoding
- binary encoding


```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df['time'].value_counts()
```

```python
# lets encode time in labelencoder usig sklearn

from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder

le = LabelEncoder()

df['encoded_time']= le.fit_transform(df['time'])
```

```python
df.head()
```

# Ordinal encoding for day column

```python
df['day'].value_counts()
```

```python
##  ordinal encoding

oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])

df['encoded_day'] = oe.fit_transform(df[['day']])

df.head()
```

### One-hot Encoder

```python
# one-hot encoder

ohe = OneHotEncoder()
ohe.fit_transform(df[['smoker']]).toarray()
```
